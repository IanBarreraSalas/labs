```python
#Author Ian Barrera Salas
plt.style.use('dark_background')

fig = plt.figure()
fig.set_dpi(100)
ax1 = fig.add_subplot(1,1,1)

#eje x
x0 = np.linspace(-3*np.pi,3*np.pi,10000)
x1 = np.linspace(-3*np.pi,3*np.pi,10000)

#tiempo
t0 = 0
t1 = 3

#ecuacipon de onda
def u(x,t):
    return 0.5*(np.cos(x+t)+np.cos(x-t))


#magnitud de ondas
x = 0.5*u(x0,t0)
y = 1*u(x1,t1)
#suma de las ondas
comb = x+y


#grafica de ondas, verde es la suma
ax1.clear()
plt.plot(x0,x,color='blue')
plt.plot(x0,y,color='yellow')
plt.plot(x0,comb,color='green')
plt.grid(True)
plt.ylim([-2,2])
 
plt.show()
```
